```python
## TODO EN S.I. ##

epsilon_0 = 8.854*(10**(-12)) #F/m
l = 0.02 #lado de la placa cuadrada
L = 0.05 #lado de la region cuadrada = 0.5
d = 0.01 #separacion de las placas
h = 0.0008 #separacion de cuadros en la malla
N = int(ceil(L/h)) #numero de puntos (impar)
V_0 = 100.0 #voltaje entre placas
V_mas = V_0/2.0 #voltaje en la placa positiva del condensador
V_menos = -V_0/2.0 #voltaje en la parte negativa del consensador
V_c = 1.0 #Voltaje constante en toda la region diferente de cero
tau = 10**(-2) #tolerancia
iteraciones = 500 #numero maximo de iteraciones

fila_abajo = (N+1)/2+int((d/2)/h) #Fila de placa inferior
fila_arriba = (N+1)/2-int((d/2)/h) #Fila de placa superior
inicio_placa = int((3*L/10)/h)
fin_placa = int((7*L/10)/h)
```

```python
print N
print fila_arriba
print fila_abajo
print inicio_placa
print fin_placa
```

```python
#Funciones que llenan el potencial

def phi(x, y, potencial):
    return (1.0/4.0)*( potencial[(x+1,y)]+potencial[(x-1,y)]+potencial[(x,y+1)]+potencial[(x,y-1)] )

def ciclo(potencial):
    
    for i in range(1,N-1):
        for j in range(1,N-1):
            if (i==fila_abajo and j>=inicio_placa and j<=fin_placa):
                potencial[(i,j)]=V_menos
            elif (i==fila_arriba and j>=inicio_placa and j<=fin_placa):
                potencial[(i,j)]=V_mas
            else:
                potencial[(i,j)]=phi(i,j,potencial)
    
    return potencial        
```

```python
#Matriz de potencial

potencial = ones((N,N))*V_c

#Condiciones de frontera de Dirichlet: en el borde de la region el potencial es cero y sobre las placas constante.

#borde
potencial[0,:] = 0.0
potencial[N-1,:] = 0.0
potencial[:,0] = 0.0
potencial[:,N-1] = 0.0

#placas
i = 3.0*L/10.0
while (i < 7.0*L/10.0):
    potencial[fila_abajo,int(i/h)]=V_menos
    potencial[fila_arriba,int(i/h)]=V_mas
    i += h
```

```python
#Ahora a hacer el campo. 

campo = zeros((N,N))
campo_y = zeros((N,N))

for i in range(0,N-1):
    for j in range(0,N-1):
        campo_y[(i,j)]=(1.0/(2.0*h))*(potencial[(i,j-1)]-potencial[(i,j+1)])
        campo[(i,j)]=(1.0/(2.0*h))*sqrt( (potencial[(i-1,j)]-potencial[(i+1,j)])**2 + (potencial[(i,j-1)]-potencial[(i,j+1)])**2)
```

```python
#Ahora la carga

n = fin_placa - inicio_placa - 2 #2 son los puntos inicial y final de la placa, donde el campo es feo
sigma = ones(n)
```

```python
for i in range(n):
    sigma[i] = epsilon_0*(campo_y[(i+inicio_placa+1,fila_arriba)] - campo_y[(i+inicio_placa+1,fila_arriba-1)])
```

```python
Q = h*l*sum(sigma)
```

```python
C = Q/V_0 #capacitancia con carga
C_0 = epsilon_0*(l**2)/d #capacitancia con formula epsilon_0*A/d
```

```python
k = 0

for j in range(0,4):
    
    d = separaciones[j]
    
    #primero el potencial
    potencial = ones((N,N))*V_c
    
    fila_abajo = (N+1)/2+int((d/2)/h) #Fila de placa inferior
    fila_arriba = (N+1)/2-int((d/2)/h) #Fila de placa superior
    inicio_placa = int((3*L/10)/h)
    fin_placa = int((7*L/10)/h)
    
    potencial[0,:] = 0.0
    potencial[N-1,:] = 0.0
    potencial[:,0] = 0.0
    potencial[:,N-1] = 0.0

    #placas
    i = 3.0*L/10.0
    while (i < 7.0*L/10.0):
        potencial[fila_abajo,int(i/h)]=V_menos
        potencial[fila_arriba,int(i/h)]=V_mas
        i += h
    
    for i in range(0,iteraciones):
        ciclo(potencial)
    
    #Ahora el campo. 
    campo_y = zeros((N,N))

    for i in range(0,N-1):
        for j in range(0,N-1):
            campo_y[(i,j)]=(1.0/h)*(potencial[(i,j)]-potencial[(i,j+1)])
    
    n = fin_placa - inicio_placa
    sigma = ones(n)
    
    for i in range(n):
        sigma[i] = epsilon_0*(campo_y[(i+inicio_placa,fila_arriba)] - campo_y[(i+inicio_placa,fila_arriba-1)])
    
    Q = h*l*sum(sigma)
    
    C = Q/V_0 #capacitancia con carga
    C_0 = epsilon_0*(l**2)/d #capacitancia con formula epsilon_0*A/d
    
    print C, C_0
    
    Cs[k]=C
    C_0s[k]=C_0
    k+=1
```
