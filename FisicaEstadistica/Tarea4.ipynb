```python
#Punto c

k = 1.38*10**(-23) #SI

def s_clasico(epsilon):
    return k*(log(epsilon)+1)

def s_cuantico(epsilon):
    return k*((epsilon+0.5)*log(epsilon+0.5) - (epsilon-0.5)*log(epsilon-0.5)) 
```

```python
epsilons = linspace(0.5001, 3, 50)

fig = figure(figsize=(16,6))

plot(epsilons, s_clasico(epsilons), c='m', label=("Clasico"))
plot(epsilons, s_cuantico(epsilons), c='g', label=("Cuantico"))

xlabel("$\epsilon / \hbar w$")
ylabel("S")

legend()
```

```python
#Punto d

def e_clasico(par):
    return par

def e_cuantico(par):
    return 0.5*(exp(1/par)+1)/(exp(1/par)-1)
```

```python
pars = linspace(0.01, 3, 50)

fig = figure(figsize=(16,6))

plot(pars, e_clasico(pars), c='m', label=("Clasico"))
plot(pars, e_cuantico(pars), c='g', label=("Cuantico"))

xlabel("$kT/\hbar w$")
ylabel("$\epsilon / \hbar w$")

legend()
```
