```python
%pylab inline
from scipy.interpolate import interp1d
```

```python
#DATA 

hidrogeno = loadtxt("./h.dat")
pxh=hidrogeno[:,0]
ih=hidrogeno[:,1]

helio = loadtxt("./he.dat")
pxhe=helio[:,0]
ihe=helio[:,1]

mercurio = loadtxt("./hg.dat")
pxhg=mercurio[:,0]
ihg=mercurio[:,1]
```

```python
m0h=argmax(ih[20:60])+19
m1h=argmax(ih[390:400])+389
m2h=argmax(ih[410:420])+409
m3h=argmax(ih[460:480])+459
m4h=argmax(ih[550:570])+549
m5h=argmax(ih[620:650])+619

m0he=argmax(ihe[0:50])
m1he=argmax(ihe[360:370])+360
m2he=argmax(ihe[370:380])+370
m3he=argmax(ihe[410:420])+410
m4he=argmax(ihe[430:450])+430
m5he=argmax(ihe[450:460])+450
m6he=argmax(ihe[460:480])+460
m7he=argmax(ihe[540:580])+540
m8he=argmax(ihe[620:640])+620
m9he=argmax(ihe[660:680])+660

m0hg=argmax(ihg[20:60])+20
m1hg=argmax(ihg[390:400])+390
m2hg=argmax(ihg[410:430])+410
m3hg=argmax(ihg[450:500])+450
m4hg=argmax(ihg[520:540])+520
m5hg=argmax(ihg[550:570])+550
```

```python
nih=zeros(len(ih))

nih[m0h]=1.5
nih[m1h]=1
nih[m2h]=1
nih[m3h]=1
nih[m4h]=1
nih[m5h]=1

nihe=zeros(len(ihe))

nihe[m0he]=1.5
nihe[m1he]=1
nihe[m2he]=1
nihe[m3he]=1
nihe[m4he]=1
nihe[m5he]=1
nihe[m6he]=1
nihe[m7he]=1
nihe[m8he]=1
nihe[m9he]=1

nihg=zeros(len(ihg))

nihg[m0hg]=1.5
nihg[m1hg]=1
nihg[m2hg]=1
nihg[m3hg]=1
nihg[m4hg]=1
nihg[m5hg]=1
```

```python
fig = figure()

plot(pxh-m0h,ih)
plot(pxhe-m0he,ihe)
plot(pxhg-m0hg,ihg)

legend(["H","He","Hg"])

xlabel("Pixel (px)")
ylabel("Intensidad (Unidades Arbitrarias)")

savefig("./espectros iniciales.png")
```

```python
#Interpolacion Cuadratica

mh=array([m0h,m1h,m2h,m3h,m4h,m5h])
lambdash=array([0, 397.0, 410.2, 434.1, 486.1, 656.3])

print mh-m0h
print ""

for i in range(len(mh)):
    print "La longitud de onda asociada a la linea",i,"del Hidrogeno es:",lambdash[i]
```

```python
interpolacion = interp1d(mh-m0h,lambdash,kind='quadratic')
```
