```python
%pylab inline
from matplotlib.pyplot import ticklabel_format
from matplotlib import ticker
```

```python
#Constantes (SI)

h = 6.62606957*10**(-34)
c = 300000000
B = 1 #Tesla
```

```python
NeAr = loadtxt("./datos/ne_ar.dat")
Ne_B_perpendicular = loadtxt("./datos/ne_b_perpendicular.dat")
Ne_B_paralelo = loadtxt("./datos/ne_b_paralelo.dat")
```

```python
fig = figure(figsize=(16,7))

x_NeAr = NeAr[:,0]
y_NeAr = NeAr[:,1]

plot(x_NeAr, y_NeAr, c='b')

xlabel("$\lambda$ (px)")
ylabel("Intensidad (Unidades Arbitrarias)")
legend(["Ne con Ar"])
xlim([200,1500])

savefig("./NeAr_px.png")
```

```python
indice_Ne = argmax(y_NeAr)
maximo_Ne = x_NeAr[indice_Ne]
intensidad_Ne = y_NeAr[indice_Ne]

indice_Ar = argmax(y_NeAr[1400:1500]) + 1400
maximo_Ar = x_NeAr[indice_Ar]
intensidad_Ar = y_NeAr[indice_Ar]

print indice_Ne, indice_Ar
print maximo_Ne, maximo_Ar
print intensidad_Ne, intensidad_Ar
```

```python
#Valores teoricos
lambda_Ne_nm = 724.5 #nm
lambda_Ar_nm = 727.2935 #nm

delta_lambda_nm = lambda_Ar_nm - lambda_Ne_nm
delta_lambda_px = maximo_Ar - maximo_Ne

print delta_lambda_nm
print delta_lambda_px
```

```python
equivalencia_px_nm = delta_lambda_nm/delta_lambda_px

#Lo que vale 1 px en nm
print equivalencia_px_nm
```

```python
cero_nm = lambda_Ne_nm - maximo_Ne*equivalencia_px_nm

#Desde donde se empieza a contar en la imagen en nm
print cero_nm
```

```python
lambdas = cero_nm + x_NeAr*equivalencia_px_nm

fig = figure(figsize=(16,7))

plot(lambdas, y_NeAr, c='b')

xlim([724.4, 727.4])
ylim([2500,34000])

xlabel("$\lambda$ (nm)")
ylabel("Intensidad (Unidades Arbitrarias)")

legend(["Ne con Ar"])

savefig("./NeAr_nm.png")
```

```python
#Para hallar la dispersion
ind_menor = where( y_NeAr[:indice_Ne] >= intensidad_Ne/2 )[0][0] - 1
ind_mayor = where( y_NeAr[indice_Ne:] <= intensidad_Ne/2 )[0][0] + indice_Ne -1 

dispersion_px = x_NeAr[ind_mayor] - x_NeAr[ind_menor]
dispersion_nm = dispersion_px*equivalencia_px_nm

print ind_menor, ind_mayor

print dispersion_px, "px"
print dispersion_nm, "nm"

print dispersion_px/(dispersion_nm*10), "px/angstrom"
```

```python
fig = figure(figsize=(16,7))

plot(x_NeAr, y_NeAr, c='b')

xlim([270,300])
ylim([2500,34000])

xlabel("$\lambda$ (px)")
ylabel("Intensidad (Unidades Arbitrarias)")

legend(["Ne con Ar"])
scatter(ind_menor, y_NeAr[ind_menor], c='r')
scatter(ind_mayor, y_NeAr[ind_mayor], c='r')

savefig("./NeAr_dispersion.png")
```

```python
fig, ax = subplots(figsize=(16,7))

y_NeAr -= amin(y_NeAr)

y_Ne_B_perpendicular = Ne_B_perpendicular[:,1]
y_Ne_B_perpendicular -= amin(y_Ne_B_perpendicular)

y_Ne_B_paralelo = Ne_B_paralelo[:,1]
y_Ne_B_paralelo -= amin(y_Ne_B_paralelo)

plot(lambdas, y_NeAr, c='b')
plot(lambdas, y_Ne_B_perpendicular, c='g')
plot(lambdas, y_Ne_B_paralelo, c='r')

legend(["Ne sin B", "Ne con B Perpendicular", "Ne con B Paralelo"])

xlim([724.4,724.6])

xlabel("$\lambda$ (nm)")
ylabel("Intensidad (unidades arbitrarias)")

ax.ticklabel_format(style='plain')
x_formatter = ticker.ScalarFormatter(useOffset=False)
ax.xaxis.set_major_formatter(x_formatter)

savefig("./Ne.png")
```
