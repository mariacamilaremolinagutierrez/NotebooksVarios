```python
%pylab inline
from matplotlib.pyplot import ticklabel_format
from matplotlib import ticker
```

```python
#Constantes (SI)

h = 6.62606957*10**(-34)
c = 300000000
B = 1 #Tesla
```

```python
NeCd = loadtxt("./datos/NeCd_186_211.dat")
Cd_B_perpendicular = loadtxt("./datos/Cd_B_Perpendicular_186_211.dat")
Cd_B_paralelo = loadtxt("./datos/Cd_B_Paralelo_186_211.dat")
```

```python
fig = figure(figsize=(16,7))

x_NeCd = NeCd[:,0]
y_NeCd = NeCd[:,1]

plot(x_NeCd, y_NeCd, c='b')

legend(["Ne con Cd"])

xlim([0,1250])
```

```python
indice_Cd = argmax(y_NeCd)
maximo_Cd = x_NeCd[indice_Cd]

indice_Ne = argmax(y_NeCd[:100])
maximo_Ne = x_NeCd[indice_Ne]
```

```python
#Valores teoricos
lambda_Cd_nm = 643.8 #nm
lambda_Ne_nm = 640.2 #nm

delta_lambda_nm = lambda_Cd_nm - lambda_Ne_nm

delta_lambda_px = maximo_Cd - maximo_Ne
```

```python
equivalencia_px_nm = delta_lambda_nm/delta_lambda_px

#Lo que vale 1 px en nm
print equivalencia_px_nm
```

```python
cero_nm = lambda_Ne_nm - indice_Ne*equivalencia_px_nm

#Desde donde se empieza a contar en la imagen en nm
print cero_nm
```

```python
lambdas = cero_nm + x_NeCd*equivalencia_px_nm

fig = figure(figsize=(16,7))

plot(lambdas, y_NeCd, c='b')

xlim([640.15, 644])
ylim([900,8500])

xlabel("$\lambda$ (nm)")
ylabel("Intensidad (unidades arbitrarias)")

legend(["Ne con Cd"])

savefig("./NeCd.png")
```

```python
#Para hallar la dispersion
maximo = amax(NeCd[1200:1250])

ind_menor = where( y_NeCd[1200:1225] >= maximo/2 )[0][0] + 1200
ind_mayor = where( y_NeCd[1223:1250] <= maximo/2 )[0][0] + 1222

dispersion_px = x_NeCd[ind_mayor] - x_NeCd[ind_menor]
dispersion_nm = dispersion_px*equivalencia_px_nm

print dispersion_px, "px"
print dispersion_nm, "nm"
```

```python
fig, ax = subplots(figsize=(16,7))

y_Cd_B_perpendicular = Cd_B_perpendicular[:,1]
y_Cd_B_paralelo = Cd_B_paralelo[:,1]

plot(lambdas, y_NeCd, c='b')
plot(lambdas, y_Cd_B_perpendicular, c='g')
plot(lambdas, y_Cd_B_paralelo, c='r')

scatter(lambdas[ind_menor], y_NeCd[ind_menor], c='red')
scatter(lambdas[ind_mayor], y_NeCd[ind_mayor], c='red')

legend(["Cd sin B", "Cd con B Perpendicular", "Cd con B Paralelo"])

xlim([643.7,643.9])

xlabel("$\lambda$ (nm)")
ylabel("Intensidad (unidades arbitrarias)")

ax.ticklabel_format(style='plain')
x_formatter = ticker.ScalarFormatter(useOffset=False)
ax.xaxis.set_major_formatter(x_formatter)

savefig("./Cd.png")
```

# $\Delta E = h \Delta f = h \Delta(\frac{c}{\lambda})$

```python
#B perpendicular

pico1_B_perpendicular_px = argmax( y_Cd_B_perpendicular[1219:1223] ) + 1219
pico2_B_perpendicular_px = argmax( y_Cd_B_perpendicular )
pico3_B_perpendicular_px = argmax( y_Cd_B_perpendicular[1232:1234] ) + 1232

pico1_B_perpendicular_nm = cero_nm + pico1_B_perpendicular_px*equivalencia_px_nm
pico2_B_perpendicular_nm = cero_nm + pico2_B_perpendicular_px*equivalencia_px_nm
pico3_B_perpendicular_nm = cero_nm + pico3_B_perpendicular_px*equivalencia_px_nm

delta_f1_perpendicular = c*( 1/(pico1_B_perpendicular_nm*10**(-9)) - 1/(pico2_B_perpendicular_nm*10**(-9)) )
delta_f2_perpendicular = c*( 1/(pico2_B_perpendicular_nm*10**(-9)) - 1/(pico3_B_perpendicular_nm*10**(-9)) )

#Delta E
delta_E1_perpendicular = h*delta_f1_perpendicular
delta_E2_perpendicular = h*delta_f2_perpendicular
```

```python
print delta_E1_perpendicular, delta_E2_perpendicular #Joules
```

```python
#B paralelo

pico1_B_paralelo_px = argmax( y_Cd_B_paralelo )
pico2_B_paralelo_px = argmax( y_Cd_B_paralelo[1228:1231] ) + 1228
pico3_B_paralelo_px = argmax( y_Cd_B_paralelo[1232:1239] ) + 1232

pico1_B_paralelo_nm = cero_nm + pico1_B_paralelo_px*equivalencia_px_nm
pico2_B_paralelo_nm = cero_nm + pico2_B_paralelo_px*equivalencia_px_nm
pico3_B_paralelo_nm = cero_nm + pico3_B_paralelo_px*equivalencia_px_nm

delta_f1_paralelo = c*( 1/(pico1_B_paralelo_nm*10**(-9)) - 1/(pico2_B_paralelo_nm*10**(-9)) )
delta_f2_paralelo = c*( 1/(pico2_B_paralelo_nm*10**(-9)) - 1/(pico3_B_paralelo_nm*10**(-9)) )

#Delta E
delta_E1_paralelo = h*delta_f1_paralelo
delta_E2_paralelo = h*delta_f2_paralelo
```

```python
print delta_E1_paralelo, delta_E2_paralelo #Joules
```

# $\Delta E = \mu_B B \quad \Rightarrow \quad \mu_B = \frac{\Delta E}{B}$

```python
magneton1_perpendicular = delta_E1_perpendicular/B
magneton2_perpendicular = delta_E2_perpendicular/B

print magneton1_perpendicular, magneton2_perpendicular
```

```python
magneton1_paralelo = delta_E1_paralelo/B
magneton2_paralelo = delta_E2_paralelo/B
 
print magneton1_paralelo, magneton2_paralelo
```

# $\mu_B = \frac{eh}{4\pi m_e c} \quad \Rightarrow \quad \frac{e}{m_e} = \frac{4\pi c \mu_B}{h}$

```python
em1_perpendicular = 4*pi*c*magneton1_perpendicular/h
em2_perpendicular = 4*pi*c*magneton2_perpendicular/h

print em1_perpendicular, em2_perpendicular
```

```python
em1_paralelo = 4*pi*c*magneton1_paralelo/h
em2_paralelo = 4*pi*c*magneton2_paralelo/h

print em1_paralelo, em2_paralelo
```
