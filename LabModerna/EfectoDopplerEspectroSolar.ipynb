```python
%pylab inline

from matplotlib.pyplot import ticklabel_format
from matplotlib import ticker
```

```python
def aproximar(x):
    if(x-int(x) < 0.5):
        return int(floor(x))
    else:
        return int(ceil(x))
    
def velocidad(separacion, lambda_cero):
    c=300000000
    z = separacion/lambda_cero
    return z,c*z
```

```python
linea_oxigeno_1_px = argmin(intensidades[1440:1490]) + 1440
linea_oxigeno_2_px = argmin(intensidades[1440:1520]) + 1440

print pixels[linea_oxigeno_1_px], pixels[linea_oxigeno_2_px]
print intensidades[linea_oxigeno_1_px], intensidades[linea_oxigeno_2_px]
```

```python
linea_oxigeno_1_nm = 630.581
linea_oxigeno_2_nm = 630.6565

equivalencia_nm_px = (linea_oxigeno_2_nm - linea_oxigeno_1_nm)/(linea_oxigeno_2_px - linea_oxigeno_1_px)

cero_nm = linea_oxigeno_1_nm - equivalencia_nm_px*linea_oxigeno_1_px
```

```python
print equivalencia_nm_px

print cero_nm
```

```python
fig = figure(figsize=(20,9))

lambdas = pixels*equivalencia_nm_px + cero_nm

xlim(626.1,630.8)
ylim(8000,34000)

xlabel('$\lambda$ (nm)', fontsize=20)
ylabel('Intensity (Arbitray Units)', fontsize=20)

plot(lambdas, intensidades)

savefig('./lineas_oxigeno_nm.png')
```
