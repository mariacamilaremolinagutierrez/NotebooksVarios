Escriba un programa que haga graficas de los lugares donde el campo electrico tiene valores constantes para dos cargas electricas $\frac{\pm q}{4\pi \epsilon_0} = 1N m^2$ ubicadas en forma de dipolo separadas una distancia $d = 3 \times 10^{−2} m$ y que oscilan con una frecuencia angular de $\omega = 2 \pi rad/s$ (usando la convencion de Griffiths para las cargas en funcion del tiempo). Estas graficas se deben hacer para los siguientes tiempos 

$t = 0$

$t = \frac{1}{4} 2 \pi \omega- 1 = \frac{\pi}{2} \omega- 1 $

$t = \frac{1}{2} 2 \pi \omega- 1 = \pi \omega- 1 $

Sabemos de clase que el campo electrico se puede escribir como:
    
##$$\vec{E} = - \frac{\mu_0 q d \omega^2}{4 \pi} \frac{\sin{\theta}}{r} \cos{\omega(t-\frac{r}{c})} \hat{\theta}$$

Simplificando y sabiendo que $\frac{q}{4\pi \epsilon_0} = 1N m^2$ :

##$$\vec{E} = - \frac{q d \omega^2}{4 \pi \epsilon_0 c^2} \frac{\sin{\theta}}{r} \cos{\omega(t-\frac{r}{c})} \hat{\theta} = - \frac{d \omega^2}{c^2} \frac{\sin{\theta}}{r} \cos{\omega(t-\frac{r}{c})} \hat{\theta}$$

```python
%pylab inline 
from math import atan2
```

```python
#Datos: (Todo en SI)

d = 3.0*10.0**(-2.0)
w = 2.0*pi
c = 3.0*10.0**8.0

ts = array([0, pi*w/2.0 - 1.0, pi*w - 1.0])
```

```python
#Función magnitud de E:

def E(r, theta, t):
    return d*(w**2)*sin(pi-theta)*cos(w*(t-r/c))/(r*(c**2))

#Función llenar matriz:

def llenar_matriz(t):
    num_xs = 1001
    num_ys = 1001

    matriz = zeros((num_xs, num_ys))

    xs = linspace(-1,1,num_xs)
    ys = linspace(-1,1,num_ys)

    for i in range(num_xs):
        for j in range(num_ys):

            x = xs[i]
            y = ys[j]

            r = (x**2 + y**2)**(0.5)

            if(r==0):
                r = (xs[i+1]**2 + ys[j+1]**2)**(0.5)

            if (x!=0):
                th = atan2(y,x)
            else:
                th = atan2(y,(xs[i+1]))

            campoE = E(r, th, t)

            matriz[i][j] = campoE
    
    return matriz
```

En coordenadas polares (este theta es en realidad pi - theta):

##$\theta = \arctan{\frac{y}{x}}$

##$r = \sqrt{x^2 + y^2}$

##$\hat{\theta} = -\sin{\theta} \hat{x} + \cos{\theta} \hat{y}$

#$t=0$:

```python
matriz = llenar_matriz(ts[0])
```

```python
fig = figure(figsize=(12,12))
imshow(matriz,extent=[-1,1,-1,1], vmin=-8*10**(-17), vmax=8*10**(-17))
colorbar()
```
