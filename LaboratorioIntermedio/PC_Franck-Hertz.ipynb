```python
#Voltaje U1 Corriente IA
#  (U1/V)     (IA/nA)

pc_175 = loadtxt('175_pc.txt')
pc_150 = loadtxt('150_pc.txt')
pc_120 = loadtxt('120_pc.txt')

x_pc_175 = pc_175[:,0]
y_pc_175 = pc_175[:,1]
x_pc_150 = pc_150[:,0]
y_pc_150 = pc_150[:,1]
x_pc_120 = pc_120[:,0]
y_pc_120 = pc_120[:,1]
```

```python
print max(x_pc_175)
print max(y_pc_175)
```

```python
figure(figsize=(16,8))

plot(x_pc_175, y_pc_175, label='$175 \quad \mathrm{K}$')
plot(x_pc_150, y_pc_150, label='$150 \quad \mathrm{K}$')
plot(x_pc_120, y_pc_120, label='$120 \quad \mathrm{K}$')

xlabel('$\mathrm{I_A(nA)}$')
ylabel('$\mathrm{U_1(V)}$')
xlim(0,61)
ylim(0,51)
legend(loc='lower right')

savefig('pc_todas.png')
```

```python
x_mins = [14, 19, 23, 28, 33, 37, 43]
x_maxs = [18, 21, 27, 31, 37, 41, 50]
x_pc_175_minimos = []

for i in range(7):
    inds = where((x_pc_175>x_mins[i]) & (x_pc_175<x_maxs[i]))
    ind_max = argmin(y_pc_175[inds])
    x_pc_175_minimos.append(x_pc_175[inds[0][ind_max]])
    
x_pc_175_minimos = array(x_pc_175_minimos)

print x_pc_175_minimos
```

```python
separaciones = []

for i in range(5):
    separaciones.append(x_pc_175_minimos[i+1]-x_pc_175_minimos[i])
    
separaciones = array(separaciones)
print separaciones
```

```python
print average(separaciones[1:])
```

```python
figure(figsize=(16,8))

plot(x_pc_175, y_pc_175)
[axvline(xm, 0, 61, c='c', linestyle='--', label=(('$%s \quad \mathrm{nA}$')%(xm))) for xm in x_pc_175_minimos]

xlabel('$\mathrm{I_A(nA)}$')
ylabel('$\mathrm{U_1(V)}$')
xlim(0,48)
ylim(0,51)
title('$\mathrm{T} = 175 \quad \mathrm{K}$')
legend(loc='upper left')

savefig('pc_175.png')
```
